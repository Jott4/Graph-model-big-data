<h1>Machine Learning - Linear Regression

ref: https://www.w3schools.com/python/python_ml_linear_regression.asp

```python
import matplotlib.pyplot as plt
#We have registered the age and speed of 13 cars as they were passing a tollbooth
#x = age // y = speed
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show() 
```

```python
import matplotlib.pyplot as plt
from scipy import stats
##para entender slope e intercept assistir https://www.youtube.com/watch?v=Ksn1g5fCe1I
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
help(stats.linregress)
```

```python
speed = myfunc(16)
print(speed)
```

a predição é que um carro de 10 anos irá passar com velocidade 85.59308314937454

<h1>Machine Learning - Polynomial Regression

The x-axis represents the hours of the day and the y-axis represents the speed:

```python
http://localhost:8888/notebooks/Machine%20Learning%20Models.ipynb#Machine-Learning---Polynomial-Regression
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show() 
```

```python
import numpy as np
from sklearn.metrics import r2_score
mymodel = np.poly1d(np.polyfit(x, y, 3))
myline = np.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
r2_score(y, mymodel(x))
```

<h1>
Machine Learning - Multiple Regression

```python
import pandas as pd
from sklearn import linear_model 
```

```python
df = pd.read_csv("cars.csv")
```

```python
df
```

```python
X = df[['Weight', 'Volume']]
```

```python
X
```

```python
y = df['CO2']
```

```python
y
```

```python
regr = linear_model.LinearRegression() #cria um modelo de linear regression
regr.fit(X, y)  #preenche com valores X e y
```

```python
help(linear_model.LinearRegression)
```

```python
#peso 2300 kg // volume 1300
predictedCO2 = regr.predict([[2300, 1300]]) 
```

```python
predictedCO2
```

Coeficiente (se x é uma variavel e 2x é 2*x, 2 é o coeficiente)

```python
regr.coef_
```

Esses valores representam os coeficientes de Peso e volume<br>
<h1>ou seja</h1><br>
a cada 1g a mais de peso, o CO2 aumenta em 0.00755095g.<br>
a cada 1 a mais de volume, o CO2 aumenta em 0.00780526.
